```python
def extract_value_from(Value):
    out = Value.replace('€', '')
    if 'M' in out:
        out = float(out.replace('M', ''))*1000000
    elif 'K' in Value:
        out = float(out.replace('K', ''))*1000
    return float(out)
```

```python
data['Wage'] = data['Wage'].apply(lambda x:extract_value_from(x) )
```

## Q6) Which Club is the Most Preferred in terms of Fame and Money

```python
data['Club'].value_counts().head(26)
```

```python
data['Value']  =data['Value'].apply(lambda x:extract_value_from(x))
```

### Real Madrid is the most popular club in terms of fame and money

```python
Popular_Club = ('Real Madrid','TSG 1899 Hoffenheim','Newcastle United','Everton','Southampton','Atlético Madrid','Chelsea',
                'RC Celta','Eintracht Frankfurt','Liverpool','Cardiff City','Wolverhampton Wanderers','AS Monaco',
                'Frosinone','Valencia CF','Rayo Vallecano','Burnley','Fortuna Düsseldorf','Arsenal','Manchester United',
                 'Manchester City','Empoli','Tottenham Hotspur','FC Barcelona','CD Leganés','Borussia Dortmund')
data_clubs = data.loc[data['Club'].isin(Popular_Club) & data['Value']]

plt.rcParams['figure.figsize'] = (20, 8)
sns.boxplot(x = data_clubs['Club'], y = data_clubs['Value'], palette = 'inferno')
plt.xticks(rotation = 90)
```
